```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count_syllables(word):
    return len(
        re.findall('(?!e$)[аоуеоыэяию]+', word, re.I) +
        re.findall('^[^аоуеоыэяию]]*e$', word, re.I)
    )
```

```python
data = pd.read_csv('Parsing/Audio/out1.csv')
```

```python
word_dict = data.to_dict()
```

```python
syllable_length = []
for i, syll in enumerate(syllable_length):
    sylls = count_syllables(word_dict['word'][i])
    if sylls == 0:
        sylls = 1
    syll = (word_dict['end'][i] - word_dict['start'][i]) / sylls
    
for i in range(data.shape[0]):
    sylls = count_syllables(word_dict['word'][i])
    if sylls == 0:
        sylls = 1
    syllable_length.append((word_dict['end'][i] - word_dict['start'][i]) / sylls)
```
